### **1. Movie Title Keywords - Worldwide Gross Income Correlation**

Adding new DF to work on so the original data will not be spoiled

```
temp_df = imdb_m.dropna(axis="index", how="any", subset=["budget", "worlwide_gross_income"])
temp_df.shape
```

Getting all unique words from titles and calculating the mean income for all unique words to find the words that made most profit.

```
pattern = r"[a-zA-Z0-9]+"
word_list = {}
for i in range(len(temp_df)):
  words = re.findall(pattern, temp_df.iloc[i].original_title)
  profit = int(temp_df.iloc[i]["worlwide_gross_income"][2:])
  for word in words:
    value_list = [profit, 1]
    if len(word) > 2 and word.lower() not in word_list:
      word_list[word.lower()] = value_list
    elif len(word) > 2:
      word_list[word.lower()][0] += profit
      word_list[word.lower()][1] += 1
meanProfit = pd.DataFrame.from_dict(word_list, orient = 'index')
meanProfit['mean_income'] = meanProfit[0] / meanProfit[1]
meanProfit.sort_values('mean_income', ascending = False, inplace = True)
meanProfit.drop(columns = [0, 1], inplace = True)
meanProfit.sort_values('mean_income', ascending = False, inplace = True)
```

Taking top 10 words

```
words = meanProfit[:10:][::-1].index.values.transpose()
profits = meanProfit[:10:][::-1].values.transpose()[0]
```

Plotting the graph

```
fig, ax = plt.subplots(figsize = (12,8))
ax.barh(words, profits, height = 0.8)
ax.set_title("Worldwide Gross Income With Respect to Words in Movie Name", size = 20)
ax.set_xlabel("Worldwide Gross Income", size = 12)
ax.set_ylabel("Words", size = 12)
xLocations = np.linspace(0, 2.5 * (10**9), 6)
xLabels = [f"$ {xLocation/(10**9):.1f}B" for xLocation in xLocations]
ax.set_xticks(xLocations)
ax.set_xticklabels(xLabels, size = 15)
ax.set_yticklabels(words, size = 15)
ax.grid(axis = "x", color = "white")
```

This graph shows the mean income value for each word in movie names. It can be seen that the words "Avatar" and "Titanic" dominated the dataset even when there is just one movie for each of those words. In addition to that, there are some franchises that have a crucial impact on data. That is why there is no correlation found and that the team has decided to exclude this data from the machine learning algorithm.

```
temp_df.head(5)
list_of_genres = temp_df['genre'].to_list()#Type: list, Each index type: str, Ex print: List of Genres:  ['Fantasy, Horror, Mystery', 'Drama', 'Drama, Romance, War',...]
#print('List of Genres: ', list_of_genres)

# # # This code finds all unique genres and adds to all_genres set # # #
#all_genres = [] # use if you want non unique genres
all_genres = set()
for genre_grp in list_of_genres:#genre_grp -> Type: string, Ex print: Adventure, Drama, Fantasy\nDrama, Romance...
  each_genre_list = genre_grp.split(", ")#each_genre_list -> Type: list, Ex print: ['Animation', 'Comedy', 'Family']\n['Drama']...
  for each_genre in each_genre_list:#each_genre -> Type: string, Ex print: ['Animation', 'Comedy', 'Family'] -> Animation\nComedy\nFamily
    #all_genres.append(each_genre)# use if you want non unique genres
    all_genres.add(each_genre) 
all_genres = list(all_genres)#typecase set to list
print(all_genres)
```

```
genres_df = pd.DataFrame(columns=['genre', 'total_no_movies_with_genre', 'genre_total_revenue', 'mean_genre_revenue'])
genres_df['genre'] = all_genres
listofzeros = [0] * len(genres_df['genre_total_revenue'])
genres_df['total_no_movies_with_genre'] = listofzeros
genres_df['genre_total_revenue'] = listofzeros
genres_df['total_no_movies_with_genre'].astype(int)
genres_df['genre_total_revenue'].astype(int)
# # #
genres_df.drop(axis = 0, labels = [14, 16], inplace = True)
temp_temp_df=temp_df

for i in range(len(temp_temp_df)):
  #profit = int(temp_temp_df.iloc[i]["worlwide_gross_income"][2:]) - makeUSD(temp_temp_df.iloc[i]["budget"]) # USELESS
  profit = int(temp_temp_df.iloc[i]["worlwide_gross_income"][2:])
  str_grp = temp_temp_df.iloc[i]["genre"].split(", ")
  common_genres = set(str_grp).intersection(all_genres)#find common genres among 2 lists
  common_genres = list(common_genres)
  for a in range(len(genres_df['genre'])):
    if  genres_df.iloc[a]["genre"] in common_genres:
      genres_df.iloc[a, 1] += 1
      genres_df.iloc[a, 2] += profit
genres_df.mean_genre_revenue = genres_df.genre_total_revenue / genres_df.total_no_movies_with_genre

display(genres_df)
```

### **4. Gender of the Cast - Worldwide Gross Income Correlation**

Data about genders of the Actor and Actress are collected and inserted in a new dataframe named new_df. After that, acttress/actor ratio is calculated for all of the movies.

```
# BERNA

#Preparing the focused dataframe
gender_df = imdb_m.merge(imdb_tp, how="outer")
gender_df.dropna(subset = ["duration","budget","avg_vote","worlwide_gross_income"], inplace = True)

gender_df = gender_df.drop(['title', 'director','year', 'date_published', 'genre','duration', 'country', 'language', 'writer', 'production_company', 'description', 
                'votes', 'usa_gross_income', 'metascore', 'ordering', 'reviews_from_users', 'reviews_from_critics', 'job', 'characters', 'avg_vote', 'actors'], axis=1)

gender_df = gender_df[(gender_df["category"] == 'actor') | (gender_df["category"] == 'actress')]

df = gender_df.groupby(by ="imdb_title_id")

#For each unique imdb_title_id --> find actress/ actor ratio

new_df = pd.DataFrame()

new_df["imdb_movie_id"] = df.groups.keys()

actors = np.zeros(len(new_df.imdb_movie_id))
actress = np.zeros(len(new_df.imdb_movie_id))
counter = 0
for i in df.groups.keys():
  for j in df.get_group(i).loc[df.get_group(i).category == "actor"].category.values:
    actors[counter] += 1
  for j in df.get_group(i).loc[df.get_group(i).category == "actress"].category.values:
    actress[counter] += 1
  counter += 1

new_df["actor_number"] = actors
new_df["actress_number"] = actress
new_df["actress/actor ratio"] = actress/actors
new_df["profit"]= df["worlwide_gross_income"]
#Find the mean profit of these each unique imdb_title_id
for i in range(len(new_df)):
  new_df.iloc[i,4] = int(new_df["profit"].iloc[i][1].values[0][2:])


new_df
```

Sorting the data and plotting the graps

```
fig, ax = plt.subplots(1, 1, figsize = (30,5))

group_list_gen = []

new_df.sort_values('actress/actor ratio', ascending = True, inplace = True)

gen_group = new_df.groupby(by="actress/actor ratio")

gen_names = gen_group.groups.keys()

new_df = new_df.replace([np.inf, -np.inf], np.nan).dropna(axis=0)

for i in gen_names:
  group_list_gen.append(gen_group.get_group(i).mean())


newlist0 = list()
for i in gen_names:
  newlist0.append(i)

actor_nums = []
actress_nums = []
ratio_nums = []
profit_nums= []

for a in range(len(group_list_gen)):#series obj
  actor_nums.append(group_list_gen[a][0])
  actress_nums.append(group_list_gen[a][1])
  ratio_nums.append(group_list_gen[a][2])
  profit_nums.append(group_list_gen[a][3])

ax.scatter(new_df["actress/actor ratio"],new_df["profit"])
ax.set_title("WorldWide Gross Income - Gender Ratio Chart")
ax.set_ylabel("WorldWide Gross Income")
ax.set_xlabel("Gender Ratio")
```

There is a correlation between the gender ratio of the cast members and this ratio’s impact on the worldwide gross income of the movie. From this graph, it can be concluded that movies which more actors took part in compared to actresses achieves higher worldwide gross income.
